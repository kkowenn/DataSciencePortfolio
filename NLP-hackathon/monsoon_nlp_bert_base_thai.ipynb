```python
import json

file_path = "/content/train.jsonl"

with open(file_path, "r") as f:
    for i, line in enumerate(f, start=1):
        try:
            json.loads(line)  # Validate each line
        except json.JSONDecodeError as e:
            print(f"Error on line {i}: {e}")
```

```python
# load dataset
train_file = "train.jsonl"
val_file = "validation.jsonl"

dataset_dict = DatasetDict({
    "train": load_dataset("json", data_files=train_file, split="train"),
    "validation": load_dataset("json", data_files=val_file, split="train"),
})
```

```python
# แก้ bug manual เลย
label_list = [
    'O', 'B_ORG', 'B_PER', 'B_LOC', 'B_MEA', 'I_DTM', 'I_ORG', 'E_ORG',
    'I_PER', 'B_TTL', 'E_PER', 'B_DES', 'E_LOC', 'B_DTM', 'B_NUM', 'I_MEA',
    'E_DTM', 'E_MEA', 'I_LOC', 'I_DES', 'E_DES', 'I_NUM', 'E_NUM', 'B_TRM',
    'B_BRN', 'I_TRM', 'E_TRM', 'I_TTL', 'I_BRN', 'E_BRN', 'E_TTL', 'B_NAME'
]
label2id = {label: idx for idx, label in enumerate(label_list)}
id2label = {idx: label for idx, label in enumerate(label_list)}
```

```python
def fix_ner_tags(example):
    """Convert ner_tags to integers and validate them."""
    example["ner_tags"] = [label2id[tag] for tag in example["ner_tags"]]
    return example


def validate_dataset(dataset_dict):
    """Ensure no invalid labels exist in the dataset."""
    for split in dataset_dict.keys():
        invalid_samples = dataset_dict[split].filter(lambda x: -1 in x["ner_tags"])
        if len(invalid_samples) > 0:
            print(f"Invalid samples found in {split}: {len(invalid_samples)}")
            raise ValueError(f"Dataset contains invalid labels in {split} split.")

dataset_dict = dataset_dict.map(fix_ner_tags)
validate_dataset(dataset_dict)
```

```python
model_checkpoint = "monsoon-nlp/bert-base-thai"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(
        examples["tokens"],
        truncation=True,
        is_split_into_words=True,
        max_length=512,
    )
    labels = []
    for i, label in enumerate(examples["ner_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        aligned_labels = []
        previous_word_idx = None
        for word_idx in word_ids:
            if word_idx is None:
                aligned_labels.append(-100)  # Ignore special tokens
            elif word_idx != previous_word_idx:
                aligned_labels.append(label[word_idx])
            else:
                aligned_labels.append(-100)  # Ignore subword tokens
            previous_word_idx = word_idx
        labels.append(aligned_labels)
    tokenized_inputs["labels"] = labels
    return tokenized_inputs

tokenized_datasets = dataset_dict.map(
    tokenize_and_align_labels,
    batched=True,
    remove_columns=["tokens", "ner_tags"],
)
```

```python
# Data Collator
data_collator = DataCollatorForTokenClassification(tokenizer)

# Output Summary
print(f"Label to ID Mapping: {label2id}")
print(f"Tokenized Dataset Example: {tokenized_datasets['train'][0]}")
print(f"Dataset Summary: {tokenized_datasets}")
```

```python
label_counts = {label: 0 for label in label_list}
for label_sequence in tokenized_datasets["train"]["labels"]:
    for label in label_sequence:
        if label != -100:
            label_counts[id2label[label]] += 1

print("Label Distribution in Training Data:", label_counts)
```

```python
# Weighted Loss
total_count = sum(label_counts.values())
weights = {label: total_count / (len(label_list) * count) for label, count in label_counts.items()}
label_weights = torch.tensor([weights[label] for label in label_list]).to("cuda" if torch.cuda.is_available() else "cpu")

```

```python
class WeightedLossTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits
        loss_fct = CrossEntropyLoss(weight=label_weights, ignore_index=-100)
        loss = loss_fct(logits.view(-1, len(label_list)), labels.view(-1))
        return (loss, outputs) if return_outputs else loss

```

```python
import numpy as np

def compute_metrics(pred):
    predictions, labels = pred
    predictions = np.argmax(predictions, axis=2)

    # Remove ignored index (-100) and flatten
    true_predictions = [
        [p for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]
    true_labels = [
        [l for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]

    true_predictions_flat = [p for sublist in true_predictions for p in sublist]
    true_labels_flat = [l for sublist in true_labels for l in sublist]

    report = classification_report(
        true_labels_flat,
        true_predictions_flat,
        labels=list(range(len(label_list))),
        target_names=label_list,
        output_dict=True,
        zero_division=0,
    )

    return {
        "macro_precision": report["macro avg"]["precision"],
        "macro_recall": report["macro avg"]["recall"],
        "macro_f1": report["macro avg"]["f1-score"],
    }

```

```python
# Define the number of labels
num_labels = len(label2id)

# Define training arguments, including the checkpoint settings
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="steps",
    eval_steps=500,
    save_steps=1000,
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    save_total_limit=2,
    report_to="none",
)
```

```python
import torch
# Load Model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = AutoModelForTokenClassification.from_pretrained(
    model_checkpoint,
    num_labels=len(label_list),
)
model.to(device)
```

```python
# Trainer
trainer = WeightedLossTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

# Train and Evaluate
trainer.train()
trainer.evaluate()
trainer.save_model("./final_model")
```

```python
import pandas as pd
import numpy as np
from transformers import Trainer

# Step 1: Load the sample submission file
submission_df = pd.read_csv("/content/sample_submission.csv")

# Debug: Check the structure of submission_df
print(f"Total IDs in submission_df: {len(submission_df)}")
print(submission_df.head(10))

# Step 2: Prepare predictions
predictions = trainer.predict(tokenized_datasets["validation"])
preds = np.argmax(predictions.predictions, axis=2)

# Step 3: Generate updates for valid tokens
submission_updates = []
for i, example in enumerate(tokenized_datasets["validation"]):
    for j, token in enumerate(example["input_ids"]):
        if token != tokenizer.pad_token_id:  # Exclude padding
            submission_updates.append((f"{i}_{j}", int(preds[i][j])))

# Create updates DataFrame
updates_df = pd.DataFrame(submission_updates, columns=["id", "ne"])

# Debug: Check the structure of updates_df
print(f"Total IDs in updates_df: {len(updates_df)}")
print(updates_df.head(10))

# Step 4: Fix the ID format in updates_df to match submission_df
updates_df["id"] = updates_df["id"].apply(lambda x: f"{int(x.split('_')[0]):05d}_{x.split('_')[1]}")

# Debug: Check the updated IDs in updates_df
print(updates_df.head(10))

# Step 5: Filter updates_df to only include IDs present in submission_df
updates_df = updates_df[updates_df["id"].isin(submission_df["id"])]

# Debug: Check the filtered updates_df
print(f"Filtered updates size: {len(updates_df)}")
print(updates_df.head(10))

# Step 6: Merge predictions into submission_df
submission_df = submission_df.merge(updates_df, on="id", how="left")

# Replace 'ne' column with the new predictions and fill missing values
submission_df["ne"] = submission_df["ne_y"].fillna(0).astype(int)  # Fill missing predictions with 0 (O label)
submission_df = submission_df.drop(columns=["ne_y"], errors="ignore")  # Drop unnecessary columns

# Debug: Check the final submission_df
print(f"Final submission_df size: {len(submission_df)}")
print(submission_df.head(10))

# Step 7: Save the final submission file
submission_df.to_csv("submission.csv", index=False)

# Debug: Ensure predictions are saved correctly
print("Submission file saved successfully!")
```
